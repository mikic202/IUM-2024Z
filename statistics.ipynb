# Artists

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
import glob
```

```python
artists = pd.read_json('data_v1/artists.jsonl', lines=True)
```

```python
geners_list = sum([gener for gener in artists['genres'].to_list() if gener is not None], [])
```

```python
unique_geners, geners_counts = np.unique(geners_list, return_counts=True)
print(unique_geners)
print(geners_counts)
print(len(unique_geners))
```

```python
unique_number_of_geners, number_of_geners_counts = np.unique(geners_counts, return_counts=True)
plt.scatter(unique_number_of_geners, number_of_geners_counts)
plt.title('Distribution of geners')
plt.xlabel('Number of artists')
plt.show()
```

# Music

```python
tracks = pd.read_json('data_v0/tracks.jsonl', lines=True)
```

```python
popularity = [track for track in tracks['popularity'].to_list() if not pd.isnull(track)]
print(popularity)
```

```python
unique_popularities, popularities_count = np.unique(popularity, return_counts=True)

plt.scatter(unique_popularities, popularities_count)
plt.title('Distribution of Popularity')
plt.xlabel('Number of artists')
plt.show()
```

# Sesions

```python
# print(glob.glob('data_v1/sesions/*.jsonl'))
entries_count = [ len(pd.read_json(file, lines=True)) for file in glob.glob('data_v1/sesions/*.jsonl')]
print(entries_count)
```

```python

mean = np.mean(entries_count)
variance = np.var(entries_count)
sigma = math.sqrt(variance)

print(f"Mean: {mean}, Variance: {variance}, Sigma: {sigma}")
print(f"Max: {max(entries_count)}, Min: {min(entries_count)}")

x = np.linspace(mean - 3*sigma, mean + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mean, sigma))

# plt.scatter(unique_entries_count, unique_entries)
plt.title('Distribution of Sesion entries')
plt.xlabel('Number of users')
plt.show()
```
